```python
#check null data
null_data=df[df.isnull().any(axis=1)]
print(null_data)
```

```python
#check null values and missing values by columnwise
df.isna()
df.isna().any()
df.isna().sum().plot(kind='bar')
df.fillna(0)
plt.show()
```

```python
df.columns
```

```python
#Total attrition of Employees by Gender
female_employees= df['Gender'].isin(['Female']).agg(sum)
male_employees= df['Gender'].isin(['Male']).agg(sum)
print(f"No of female employees: {female_employees}\nNo of male employees: {male_employees}")
```

```python
#Tootal number of Employees
total_employees=female_employees+male_employees
print(f"Total No of  employees:{total_employees}")
```

```python
#Total attrition of Employees by Gender
colors = {'Male': 'blue', 'Female': 'pink'}
sns.countplot(x='Gender',data=df, hue='Gender' , hue_order=['Male','Female'], palette=colors)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Total attrition of Employees by Gender')
```

```python
fig=px.bar(df.groupby('Gender')['EmployeeCount'].sum(), text=df.groupby('Gender')['EmployeeCount'].sum(), labels={'value':'No of Employees'})
fig.update_layout(width=400, height=300, title='Gender')
fig.show()
```

```python
#Total attrition by Department
employee_group_department=df.groupby(['Department'])['EmployeeCount'].sum()
print(type(employee_group_department))
labels=['HR','R&D','Sales']
employee_group_department.plot(kind='pie',labels=labels, autopct='%1.1f%%',startangle=90)
plt.legend(labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
employee_group_department.plot(kind='bar')
```

```python
#Total attrition by age
age_bins = [18, 31, 46, 61]
age_labels = ['18-30', '31-45', '46-60']

df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Grouped bar plot for total attrition by age group
sns.countplot(x='AgeGroup', data=df, order=age_labels, hue='AgeGroup', palette='muted')

plt.xlabel('Age Group')
plt.ylabel('Employee Count')
plt.title('Total Attrition by Age Group')
plt.show()
```

```python
#Total attrition by distancefromhome
distance_bins = [1, 11, 21, 31]
distance_labels =['1-10 (Nearby)', '11-20 (Far)', '21-30 (Far Away)']

df['DistanceGroup'] = pd.cut(df['DistanceFromHome'], bins=distance_bins, labels=distance_labels, right=False)


# Grouped bar plot for total attrition by distancefromhome
sns.countplot(x='DistanceGroup', data=df, order=distance_labels, hue='DistanceGroup', palette='muted')

plt.xlabel('Distance From Home by km')
plt.ylabel('Employee Count')
plt.title('Total Attrition by Distancefromhome')
plt.show()
```

```python
#Total attrition by work life balance
mapping = {1: 'Bad', 2: 'Average', 3: 'Satisfactory', 4: 'Excellent'}

# Map numerical values to categories in a new column 'WorkLifeBalanceCategory'
df['WorkLifeBalanceCategory'] = df['WorkLifeBalance'].map(mapping)

# Count the number of employees for each WorkLifeBalanceCategory value
employee_group_worklifebalance = df.groupby('WorkLifeBalanceCategory')['WorkLifeBalanceCategory'].count().reset_index(name='EmployeeCount')

# Bar plot with color based on WorkLifeBalanceCategory
sns.barplot(x='WorkLifeBalanceCategory', y='EmployeeCount', data=employee_group_worklifebalance, hue='WorkLifeBalanceCategory', palette='muted', order=['Bad', 'Average', 'Satisfactory', 'Excellent'])
plt.xlabel('Employee WorkLifeBalance')
plt.title('Total attrition by Work Life Balance')
```

```python
#Total attrition by marital status
employee_group_maritalstatus = df.groupby(['MaritalStatus', 'Gender']).size().unstack()
print(employee_group_maritalstatus)
ax = employee_group_maritalstatus.plot(kind='bar', stacked=False)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.xticks(rotation=45)
plt.title('Total attrition by marital status')
```

```python
#Total attririon by Education field
sns.countplot(data=df, x='EducationField' , hue='EducationField' , order=['Life Sciences', 'Medical', 'Marketing',
       'Technical Degree', 'Human Resources','Other'] , palette='dark')
plt.xlabel('Education Field')
plt.ylabel('Number of Employees')
plt.title('Total attririon by Education field')
plt.xticks(rotation=45)
```

```python
#Total attririon by Job Role
groupbyjobrole=df.groupby(['JobRole'])['JobRole'].count()
print(groupbyjobrole)
sns.countplot(x=df['JobRole'], hue=df['JobRole'])
plt.xticks(rotation=80)
plt.ylabel('Number of Employees')
plt.title('Total attririon by Job Role')
fig_groupbyjobrole=px.bar(groupbyjobrole , text=groupbyjobrole)
fig_groupbyjobrole.show()
```
